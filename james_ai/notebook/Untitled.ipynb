```python
words = open('../formatted_files/three_letter_words.txt').read().splitlines()
```

```python
b = {}
for word in words:
    chs = ['$'] + list(word) + ['$']
    for ch1, ch2 in zip(word, word[1:]):
        bigram = (ch1, ch2)
        b[bigram] = b.get(bigram, 0) + 1
```

```python
sorted(b.items(), key = lambda kv: -kv[1])
```
